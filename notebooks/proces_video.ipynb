```python
def divide_into_parts(total, x):
    # Calculate the number of parts
    if x <= 0:
        raise ValueError("The size of each part must be greater than zero.")
    
    num_parts = total // x
    remainder = total % x

    remainder_for_each_part = remainder

    # Create the parts
    parts = [x] * num_parts

    # Distribute the remainder evenly
    for i in range(len(parts)-1):
        parts[i] += 1  # Append 1 to avoid index out of range error

    # If there are no parts yet, ensure we at least have one part
    if not parts:
        parts.append(total)

    return parts

# Example usage:
total = 101
x = 15
parts = divide_into_parts(total, x)
print(parts)
```

```python
import numpy as np
def divide_into_parts(total_number, size):
    parts = None
    remainder = 1
    num_parts = 0
    while remainder > num_parts:

        if parts is None:
            num_parts = int(total_number // size)
            remainder = total_number - num_parts * size
            remainder_decimals = remainder - int(remainder)
            remainder = int(remainder)
            parts = [size] * num_parts
            parts = np.array(parts, dtype="float")
        else:
            remainder_per_part = remainder // num_parts
            remainder = remainder % num_parts

            parts += remainder_per_part

    parts[:remainder] += 1
    parts[remainder] += remainder_decimals

    return parts
```
